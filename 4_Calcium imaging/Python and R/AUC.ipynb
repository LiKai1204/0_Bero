```python
#Plotting with Python
import numpy as  np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import os
```

```python
os.chdir( "/Users" )

sns.set()
sns.set_style('white')
sns.set_palette('Set1')
```

```python
df_C = pd.read_csv(' ', header=None)
df_B = pd.read_csv(' ', header=None)
```

```python
df_C = df_C * 249
df_B = df_B * 249
```

```python
df_C_melt = pd.melt(df_C)
df_C_melt['conditions'] = 'Control'

df_B_melt = pd.melt(df_B)
df_B_melt['conditions'] = 'beroKD'
```

```python
df = pd.concat([df_C_melt , df_B_melt], axis=0)
```

```python
fig = plt.figure(figsize=(5,10))
ax = fig.add_subplot(1, 1, 1)
sns.boxplot(x='conditions', y='value', data=df, showfliers=False, ax=ax, )
sns.stripplot(x='conditions', y='value', data=df, dodge=True, jitter=True, color='black', size=15, ax=ax)

ax.set_xlabel('')
ax.set_ylabel('')
ax.set_yticks([0, 500, 1000, 1500, 2000, 2500, 3000])
plt.grid(True)
```

```python
plt.show()
fig.savefig(" ")
```
